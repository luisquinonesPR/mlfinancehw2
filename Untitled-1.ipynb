```python
import yfinance as yf
import pandas as pd

start = '2015-01-01' 
end = '2020-06-01'

stock_namelist = ['AAPL', 'ABBV', 'AMZN', 'DB', 'DIS', 'GOOG', 'HAL', 'HSBC', 'JPM', 'KO', 'MCD', 'MSFT', 'PFE', 'XOM']
sector_namelist = ['Technology', 'Healthcare', 'Technology', 'Finance', 'Entertainment', 'Technology', 'Energy', 'Finance', 'Finance', 'Consumer', 'Consumer', 'Technology', 'Healthcare', 'Energy']

prices = pd.DataFrame()

for stock_index in range(len(stock_namelist)):
    stock_data = yf.download(stock_namelist[stock_index], start=start, end=end, interval='1d')
    prices[stock_namelist[stock_index]] = stock_data['Adj Close']

prices.index = pd.to_datetime(prices.index)

print(prices)
```

```python
import numpy as np
X = np.diff(np.log(prices.values), axis=0)[1:]
N = X.shape[1]  # number of stocks
T = X.shape[0]  # number of days
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Compute normalized prices
normalized_prices = prices / prices.iloc[0, :]
N = normalized_prices.shape[1]  # number of stocks

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, gridspec_kw={'height_ratios': [7, 1]})

# Plot normalized prices in the first subplot
for i in range(N):
    ax1.plot(normalized_prices.iloc[:, i], label=prices.columns[i])
ax1.set_title('Normalized prices')

# Remove margins and ticks from the legend subplot
ax2.axis('off')

# Add legend to the legend subplot
legend = ax2.legend(prices.columns, loc='center', ncol=np.ceil(N/2), frameon=False, fontsize=7)

plt.tight_layout()
plt.show()
```

```python
import cvxpy as cp

def portfolio_max_return_robust_ellipsoid(mu_hat, S, kappa=0.1):
    N = len(mu_hat)
    w = cp.Variable(N)
    S12 = np.linalg.cholesky(S)
    prob = cp.Problem(
        cp.Maximize(mu_hat @ w - kappa * cp.norm2(S12 @ w)),
        [w >= 0, cp.sum(w) == 1]
    )
    prob.solve()
    return w.value

weights = portfolio_max_return_robust_ellipsoid(mu, Sigma)

print(weights)
```
